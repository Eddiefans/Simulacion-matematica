## Integración Montecarlo

1. Realizar las siguientes integrales. Poner los resultados en una tabla cuyas filas correspondan a la cantidad de términos utilizados en la aproximación (usar 10, 100, 1000, 10000 y 100000 términos) y cuyas columnas correspondan a las funciones.
    + $\int_{1}^{3} e^{x}\log(x)\text{d}x$.
    + $\int_1^3 \sin(x)\ln(x) \text{d}x$

```python
import numpy as np
import pandas as pd

def montecarlo(f, N, a, b):
    u = np.random.uniform(a, b, N)
    return np.mean(f(u))*(b-a)

def f1(x):
    return np.exp(x)*np.log10(x)

def f2(x):
    return np.sin(x)*np.log(x)

N = np.logspace(1, 5, 5)
resul = [6.09207, 0.869844]


table = pd.DataFrame(columns=["Aproximación 1", "Error 1", "Aproximación 2", "Error 2"], index=N)

for i in N:
    table.loc[i, "Aproximación 1"] = montecarlo(f1, int(i), 1, 3)
    table.loc[i, "Aproximación 2"] = montecarlo(f2, int(i), 1, 3)
table["Error 1"] = np.abs(resul[0] - table["Aproximación 1"])     
table["Error 2"] = np.abs(resul[1] - table["Aproximación 2"])  

print(table)
```

$$x^2+2y^2\leq-2x+8y-7$$

Poner los resultados en una tabla cuyas filas correspondan a la cantidad de términos utilizados en la aproximación (usar 10, 100, 1000, 10000 y 100000 términos) y en las columnas el valor de aproximación y el error relativo.

Definir unos límites adecuados $a_1$, $b_1$, $a_2$ y $b_2$, para la región $R=\left[a_1,b_1\right]\times\left[a_2,b_2\right]$

```python
x = np.random.uniform(-2.5, 0.5, 100)
y = np.random.uniform(0.9, 3.1, 100)
X, Y = np.meshgrid(x, y)
pt.figure()
pt.scatter(X, Y, c=~bernoulli(X, Y))
```

```python
import matplotlib.pyplot as pt
import pandas as pd
import numpy as np
%matplotlib inline

def montecarlo2(region, ax, bx, ay, by, N):
    x = np.random.uniform(ax, bx, int(N))
    y = np.random.uniform(ay, by, int(N))
    B = region(x, y)
    AR = (bx - ax)*(by - ay)
    error = np.sqrt((1-np.mean(B))/(np.mean(B)))
    return error, np.mean(B)*AR
def bernoulli(x, y):
    return x**2+2*y**2 <= -2*x+8*y-7

N = np.logspace(1, 5, 5)
table = pd.DataFrame(columns=["Aproximación", "Error"], index=N)

for i in N:
    table.loc[i, "Aproximación"] = montecarlo2(bernoulli, -2.5, 0.5, 0.9, 3.1, i)[1]
    table.loc[i, "Error"] = montecarlo2(bernoulli, -2.5, 0.5, 0.9, 3.1, i)[0]
print(table)
```

3. Describe 3 casos de uso de integración Montecarlo en tu ingeniería.

Opciones europeas.

Por ejempo, al tener una opción put con los siguientes datos:
- Duración de 6 meses
- Precio inicial de 50 
- Precio de ejercio de 52
- Volatilidad del subyacente de 12% anual 
- Tasa de interés de 6% anual

El valor de la opción está dado por:
$V_{t} = E_{*}(e^{-r(T-t)}h(S_{T})|F_{t})$

Por lo tanto, podemos usar montecarlo para realizar la integral y saber una aproximación del valor del put.

Referencia: http://mat.izt.uam.mx/mat/documentos/notas%20de%20clase/cfenaoe3.pdf

Rentabilidad de proyectos.

Dentro de proyectos, es necesario saber el TIR y VAN, tasa interna de retorno y valor acutal neta, de esta manera podremos saber el valor de la empresa y recuperación de inversión aproximados. 

Al ser valores estimados, el método montecarlo ayuda mucho a tener unas aproximaciones más acertadas.

Riesgos.

Para saber los riesgos de una inversión, se toma en cuenta la probabilidad de que ocurra un evento y la gravedad del mismo, esto es el capital o valor de nuestra inversión.

Para tener una aproximación con un grado de confianza más alto, se puede utilizar el método Montecarlo para hacer estimaciones.
