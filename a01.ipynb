```python
import pandas as p

print("Lista de 100 multiplos de 3: ")
list3 = []
for i in range(1, 101, 1):
    list3.append(i*3)
print(list3)


print("IMC:")
dataPE = p.read_excel("datosPE.xlsx")
listIMC = []
for i in range(0, len(dataPE)):
    listIMC.append({"Id":i+1, "IMC":round(dataPE["Peso"][i]/(dataPE["Estatura"][i]**2), 2)})
    print(listIMC[-1])


dataCT = p.read_excel("datosCT.xlsx")
listProfit = []
listNegProfit = []
totalIncome = 0
print("Utilidades:")
for i in range(0, len(dataCT)):
    listProfit.append({"Mes": dataCT["Unnamed: 0"][i], "Utilidad": dataCT["Ingreso"][i]-dataCT["Gasto"][i]})
    print(listProfit[-1])
    if dataCT["Ingreso"][i]-dataCT["Gasto"][i] < 0:
        listNegProfit.append(dataCT["Gasto"][i])
    totalIncome += dataCT["Ingreso"][i]
print("Promedio de gastos de meses con utilidades negativas: {}".format(sum(listNegProfit)/len(listNegProfit)))
print("Total da ingresos del año: {}".format(totalIncome))

```
