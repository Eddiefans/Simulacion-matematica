### Decisiones de inversión usando ecuaciones diferenciales

**Objetivo General:**

El objetivo de este proyecto es mostrar una simulación que pueda ayudar a alguna persona encontrar un plan de inversión y su ganancia real.

Para esto se utilizarán ecuaciones diferenciales, además de una simulación de esta para hacer el proceso más simple y automatizado.


**Objetivo Específico:**

Hay una persona que quiere contratar un plan de ahorro. 
Tiene dos opciones con el mismo capital a invertir y el mismo periodo. Por lo tanto, no entiende muy bien la diferencia entre los dos y desea saber cuál es mejor opción.

Opción A:
- Abonos de $30,000.00 semestrales.
- Tasa de interés de 8.7% anual.
- Duración de 10 años.

Opción A:
- Abonos de $2,500.00 quincenales.
- Tasa de interés de 0.8% mensual.
- Duración de 10 años.

Ambas tasas son compuestas continuamente.
El objetivo es encontrar la mejor opción, con mayor ganancia y poder explicar la razón de esto.


### Modelo que representa el problema

**Ecuaciones diferenciales**

Para la solución de una inversión continua, podemos usar ecuaciones diferenciales, ya que buscamos el cambio de nuestra inversión con respecto del tiempo.

Si sabemos que el cambio de nuestro capital es igual a los intereses (capital actual por tasa) más los abonos, tenemos lo siguiente:

$\frac{dS}{dt} = rS + A$

- r = tasa de interés
- S = capital 
- A = abono

Específicamente para:

**Opción A:**

$r = 8.7\%/2 = 4.35\%$

$\frac{dS}{dt} = 4.35\%S + 30,000$


**Opción B:**

$r = 0.8\%/2 = 0.4\%$

$\frac{dS}{dt} = 0.4\%S + 2,500$


### Solución y simulación del problema

**Solución analítica**

Teniendo la forma general de una ecuación diferencial lineal de primer orden:

$\frac{dy}{dx} + p(x)y = q(x)$

y teniendo:

$\frac{dS}{dt} - rS = A$

sabemos que:
- x = t
- y = S
- p(x) = -r
- q(x) = A

Si la formula nos dice que $\mu = e^{\int p(x)dx}$ entonces $\mu = e^{\int -rdx}$

Y $\mu = e^{-rx}$

La fórmula para resolver una ecuación de primer orden es:

$y\mu = \int q\mu dx$

Por lo tanto:

$-ye^{-rx} = \int -Ae^{-rx} dx$

Usando integración por partes, queda que:

$ye^{-rx} = Ae^{-rx}\frac{1}{r} + C$

Despejando para y:

$y = -\frac{A}{r} + Ce^{rx}$

Sustituyendo los valores originales, tenemos que:

$S = -\frac{A}{r} + Ce^{rt}$

Como se sabe que el capital es igual a 0 cuando el tiempo es 0, tenemos que:

$0 = -\frac{A}{r} + C*1$

$C = \frac{A}{r}$

Por lo tanto:

$S = -\frac{A}{r} + \frac{A}{r}e^{rt}$

$S = \frac{A}{r}(e^{rt}-1)$


```python
import numpy as np
import matplotlib.pyplot as pt
%matplotlib inline
from scipy.integrate import odeint
```

```python
def solucion(t, r, A):
    return (A/r)*(np.exp(r*t)-1)

def solucion_num(S, t, r, A):
    return r*S + A

# Opción A
ra = 0.087/2
Aa = 30000
ta = np.linspace(0, 10*2, 10*24 + 1)

solucionA_a = solucion(ta, ra, Aa)
solucionA_n = odeint(solucion_num, 0, ta, args=(ra, Aa))


pt.figure()
pt.title("Solución numérica vs analítica en caso A")
pt.plot(ta, solucionA_n, lw=5, label="Númerica")
pt.plot(ta, solucionA_n, label="Analítica")
pt.legend()
pt.grid()

# Opción B
rb = 0.008/2
Ab = 2500
tb = np.linspace(0, 10*24, 10*24 + 1)

solucionB_a = solucion(tb, rb, Ab)
solucionB_n = odeint(solucion_num, 0, tb, args=(rb, Ab))

pt.figure()
pt.title("Solución numérica vs analítica en caso B")
pt.plot(tb, solucionB_n, lw=5, label="Númerica")
pt.plot(tb, solucionB_n, label="Analítica")
pt.grid()
pt.legend()
```

### Visualización de la solución del problema

```python
t = ta = np.linspace(0, 10, 10*24 + 1)
pt.figure()
pt.title("Comparación de las opciones")
pt.plot(t, solucionA_n, label="Opción A")
pt.plot(t, solucionB_n, label="Opción B")
pt.grid()
pt.legend()
print("El capital final de la opción A es de: {}".format(solucionA_n[-1][0]))
print("El capital final de la opción B es de: {}".format(solucionB_n[-1][0]))
```

Como podemos ver en la gráfica y resultados:

La opción B es mejor que la opción A

Por lo tanto, es mejor invertir en la opción B, al menos que la persona vea necesario el abono semestral.
